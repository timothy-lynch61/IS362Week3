__Calculating Mileage for the Day__

For this we will need to create a function that subtracts the current days mileage from the previous days, to get how much we have traveled today. Down below dataSeries refers to day_miles_series, and that should be used to call the function.

Remember that Day 1 will be 0 in the index and therefore to search for Day 2 you would actual enter in the number 1. To search for the first day you would use the number 0 for the second parameter whichDay.

```python
def subtract(dataSeries, whichDay):
    if whichDay == 0:
        return dataSeries[0]
    else:
        return dataSeries[whichDay] - dataSeries[whichDay - 1]
    
subtract(day_miles_series, 2)
```
